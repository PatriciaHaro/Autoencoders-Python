```python
mesh = trimesh.load(os.path.join(DATA_DIR, "chair/train/chair_0021.off"))
mesh.show()
```

```python
points = mesh.sample(2048)

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(points[:, 0], points[:, 1], points[:, 2])
ax.set_axis_off()

plt.show()
```

```python
voxel = trimesh.voxel.creation.voxelize(mesh=mesh, pitch=1, method='subdivide')
voxel.show()
```
